```python
import pandas as pd
import numpy as np
import random
```

```python
morning_path = ".\ML-Capstone-main\Morning_Meal.csv"
afternoon_path = ".\ML-Capstone-main\Afternoon_Meals.csv"
night_path = ".\ML-Capstone-main\\Night_Meals.csv"

morning_meals = pd.read_csv(morning_path, on_bad_lines="skip")
afternoon_meals = pd.read_csv(afternoon_path, on_bad_lines="skip")
night_meals = pd.read_csv(night_path, on_bad_lines="skip")

```

```python
# Function Recommendation Each time in a day
def each_meals_rec(day_meals):
    
    food_names = day_meals["Nama"]
    # Shuffle the name
    shuffled_food = food_names.sample(frac=1) 
    #Display few food
    choices_food = shuffled_food[:10]
    
    return day_meals[day_meals["Nama"].isin(choices_food)]
```

```python
#Recommendation Morning Meals
each_meals_rec(morning_meals)
```

```python
#Recommendation afternoon Meals
each_meals_rec(afternoon_meals)
```

```python
#Recommendation Night Meals
each_meals_rec(night_meals)
```
