```python
import pandas as pd
import numpy as np
```

```python
path_data = ".\ML-Capstone-main\Training_result.csv"
path_morning = ".\ML-Capstone-main\Morning_Meal.csv"
path_afternoon = ".\ML-Capstone-main\Afternoon_Meals.csv"
path_night = ".\\ML-Capstone-main\\Night_Meals.csv"

dataset = pd.read_csv(path_data, on_bad_lines="skip")
morning_data = pd.read_csv(path_morning, on_bad_lines="skip")
afternoon_data = pd.read_csv(path_afternoon, on_bad_lines="skip")
night_data = pd.read_csv(path_night, on_bad_lines="skip")
dataset
```

```python
# Convert the values in the Name column to lowercase
dataset.iloc[:, 0] = dataset.iloc[:, 0].str.lower()
morning_data.iloc[:, 0] = morning_data.iloc[:, 0].str.lower()
afternoon_data.iloc[:, 0] = afternoon_data.iloc[:, 0].str.lower()
night_data.iloc[:, 0] = night_data.iloc[:, 0].str.lower()
```

```python
#Take only food names
food_name = dataset['Nama'].values
morning_meals = morning_data['Nama'].values
afternoon_meals = afternoon_data['Nama'].values
night_meals = night_data['Nama'].values
```

```python
#FIltering from food datasets into three meals each day
morning_meals_nutrition = []
afternoon_meals_nutrition = []
night_meals_nutrition = []

for i in food_name :
    if i in morning_meals :
        morning_meals_nutrition.append(i)

for i in food_name :
    if i in afternoon_meals :
        afternoon_meals_nutrition.append(i)
        
for i in food_name :
    if i in night_meals :
        night_meals_nutrition.append(i)

data_morning = dataset[dataset["Nama"].isin(morning_meals_nutrition)]
data_afternoon = dataset[dataset["Nama"].isin(afternoon_meals_nutrition)]
data_night = dataset[dataset["Nama"].isin(night_meals_nutrition)]
```

```python
#Save Again 
data_morning.to_csv(".\ML-Capstone-main\Morning_Meal.csv", index=False)
data_afternoon.to_csv(".\ML-Capstone-main\Afternoon_Meals.csv", index=False)
data_night.to_csv(".\\ML-Capstone-main\\Night_Meals.csv", index=False)
```
